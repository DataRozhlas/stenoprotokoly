```python
import polars as pl
import altair as alt
```

```python
pl.Config.set_tbl_rows(50)
pl.Config.set_fmt_str_lengths(150)
pl.Config.set_tbl_width_chars(200)
```

```python
slovenske = [
    "S SK 1939",
    "S SR 1939-1945"
]
```

```python
df = pl.read_parquet("data_raw/smazat_projevy.parquet").sort(by='datum').with_columns(
    pl.col('text').str.len_bytes().alias('delka')
).filter(
    ~pl.col('komora_komplet').str.contains('SNR')
).filter(
    ~pl.col('komora_komplet').str.contains('SNR')
).filter(
    ~pl.col('komora_komplet').str.contains('NR SR')
).filter(
    ~pl.col('komora_komplet').is_in(slovenske)
)
```

```python
delka_filtru = 500
vse = df.filter(~pl.col('datum').is_null()).filter(pl.col('delka') > delka_filtru).group_by_dynamic(index_column="datum", every="1y").agg(pl.col('text').len())
```

```python
alt.Chart(
    df.filter(~pl.col('datum').is_null()).filter(pl.col('delka') > delka_filtru).group_by_dynamic(index_column="datum", every="1y").agg(pl.col('text').len()).to_pandas(),
    width=1000
).mark_bar(
).encode(
    alt.X("datum:T"),
    alt.Y("text:Q")
)
```

```python
def vyvoj(slovo, rok=0):
    return df.filter(~pl.col('datum').is_null()).filter(pl.col("datum").dt.year() >= rok).filter(pl.col('text').str.contains(slovo)).filter(pl.col('delka') > delka_filtru).group_by_dynamic(index_column="datum", every="1y").agg(pl.col('text').len()).join(
        vse, on='datum',how='left'
    ).with_columns((pl.col('text') / pl.col('text_right')).alias('podil'))
```

```python
def graf(frejm):
    return alt.Chart(
    frejm.filter(pl.col('datum').dt.year() >= 1918).to_pandas(), 
        width=500, height=300
).mark_bar().encode(
    alt.X("datum:T"),
    alt.Y("podil:Q")
)
```

```python
def porovnani(slovnik: dict):
    
    frejm = pl.concat([vyvoj(regular, rok=1918).with_columns(pl.lit(popis).alias('co')) for popis, regular in slovnik.items()])
    
    return alt.Chart(
        frejm.to_pandas()
    ).mark_line(
    ).encode(
        alt.X('datum:T'),
        alt.Y('podil:Q'),
        alt.Color('co:N')
    )
```
