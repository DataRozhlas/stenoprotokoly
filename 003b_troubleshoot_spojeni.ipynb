Odhaluje konkrétní soubory s metadaty o přepisech, které se odlišují od schématu. Obvykle není důvod spouštět.

```python
import polars as pl
import glob
from pathlib import Path

# Get list of all CSV files
csv_files = glob.glob('data_raw/schuze/meta_*.csv')

schema = pl.Schema({
    'komora': pl.String,
    'obdobi': pl.Int32,
    'schuze': pl.Int32,
    'cast': pl.Int32,
    'soubor': pl.String,
    'datum': pl.String,
    'komora_komplet': pl.String,
    'schuze_komplet': pl.String,
    'prepsano': pl.Boolean,
    'autorizovano': pl.Boolean
})

problematic_files = []

print("jedu!")

for file_path in csv_files:
    try:
        df = pl.read_csv(file_path, schema=schema)
        # Try the date conversion
        df_converted = df.with_columns(
            pl.col("datum").str.to_date(format="%Y-%m-%d", exact=False)
        )

    except Exception as e:
        print(f"✗ {file_path} - ERROR: {e}")
        problematic_files.append(file_path)

print(f"\nProblematic files: {problematic_files}")
```
