```python
import altair as alt
```

```python
alt.Chart(
    df.filter(pl.col('datum').is_not_null()).sort(by='datum').group_by_dynamic(index_column='datum',every='1y').agg(pl.col('delka_textu').sum() / 1800).to_pandas(),
    width=800
).mark_bar().encode(
    alt.X('datum:T'),alt.Y('delka_textu:Q')
)
```

```python
df.filter(pl.col('datum').is_not_null()).sort(by='datum').group_by_dynamic(index_column='datum',every='1y').agg(pl.col('delka_textu').sum() / 1800).tail(20)
```

```python
df.columns
```

```python
df.filter(pl.col('mluvci').str.contains("Hlasování")).group_by(["mluvci",'soubor']).len().sort(by="len",descending=True)
```

```python
kontrola_delky = df.filter(pl.col('datum').is_not_null()).sort(by='datum').group_by_dynamic(index_column='datum',every='1y').agg(pl.col('delka_textu').sum()).join(
    df.filter(pl.col('datum').is_not_null()).sort(by='datum').with_columns((pl.col('soubor') + pl.col('mluvci') + pl.col('text')).alias('id')).group_by_dynamic(index_column='datum',every='1y').agg(pl.col('id').unique().len()),
    how='left', on='datum'
).with_columns((pl.col('delka_textu') / pl.col('id')).alias('prumerna_delka_projevu'))
kontrola_delky
```

```python
df.filter(pl.col("text").str.contains("Dalším řečníkem je pan posl. Richter. Uděluji mu slovo."))
```

```python
alt.Chart(
    kontrola_delky.to_pandas(),
    width=800
).mark_bar().encode(
    alt.X('datum:T'),
    alt.Y('prumerna_delka_projevu:Q')
)
```

```python
alt.Chart(
    kontrola_delky.filter(pl.col('datum').dt.year() >= 1918).to_pandas(),
    width=800
).mark_bar().encode(
    alt.X('datum:T'),
    alt.Y('prumerna_delka_projevu:Q')
)
```
