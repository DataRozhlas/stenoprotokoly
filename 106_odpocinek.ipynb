```python
import polars as pl
import altair as alt
```

```python
pl.Config.set_tbl_rows(150)
pl.Config.set_fmt_str_lengths(150)
pl.Config.set_tbl_width_chars(200)
```

```python
df = pl.read_parquet(
    "data/projevy.parquet"
).filter(
    pl.col('datum').dt.year() >= 1918
).drop_nulls(
    subset=['datum']
)
```

```python
def hledej(retezec, doslova=False, frejm=df):
    return frejm.filter(
    pl.col('text').str.contains(retezec, literal=doslova)
).group_by_dynamic(
    index_column='datum',every='1y').agg(pl.col('text').unique().len()
).join(
    frejm.group_by_dynamic(index_column='datum',every='1y').agg(pl.col('text').unique().len()), 
    on='datum', how='left'
).with_columns(
    (pl.col('text') / pl.col('text_right')).alias('podil')
).with_columns(
        pl.lit(retezec).alias('co')
)
```

```python
alt.Chart(
    hledej("[Vv]oln\w{1,4} čas").to_pandas(),
    width = 600
).mark_bar(
).encode(
    alt.X('datum:T'),
    alt.Y('podil:Q')
)
```

```python
import re
```

```python
def tydny_dovolene(retezec):
    try:
        return re.search(r'\w{1,10} týd\w{2,10} dovolené', retezec).group()
    except:
        return None
```

```python
tydny_dovolene('8 týdnů dovolené')
```

```python
df = df.with_columns(pl.col('text').map_elements(tydny_dovolene, return_dtype=pl.String).alias('dovolena'))
```

```python
df.with_columns(pl.col('dovolena').str.to_lowercase()).group_by('dovolena').len().sort(by='len',descending=True)
```
