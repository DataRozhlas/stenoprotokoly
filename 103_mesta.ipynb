```python
import polars as pl
import altair as alt
```

```python
pl.Config.set_tbl_rows(150)
pl.Config.set_fmt_str_lengths(150)
pl.Config.set_tbl_width_chars(200)
```

```python
df = pl.read_parquet(
    "data/projevy.parquet"
).filter(
    pl.col('datum').dt.year() >= 1918
).drop_nulls(
    subset=['datum']
)
```

```python
snemovny = df.filter(~pl.col('komora_komplet').str.contains('Senát')).filter(~pl.col('komora_komplet').str.contains('Společ')).select(pl.col('komora_komplet')).unique().to_series().to_list()
```

```python
df = df.filter(pl.col("komora_komplet").is_in(snemovny)).sort(by="datum")
```

```python
def hledej(retezec, doslova=False, frejm=df):
    return frejm.filter(
    pl.col('text').str.contains(retezec, literal=doslova)
).group_by_dynamic(
    index_column='datum',every='1y').agg(pl.col('text').unique().len()
).join(
    frejm.group_by_dynamic(index_column='datum',every='1y').agg(pl.col('text').unique().len()), 
    on='datum', how='left'
).with_columns(
    (pl.col('text') / pl.col('text_right')).alias('podil')
).with_columns(
        pl.lit(retezec).alias('co')
)
```

```python
alt.Chart(
    hledej("Brno").to_pandas(),
    width = 600
).mark_bar(
).encode(
    alt.X('datum:T'),
    alt.Y('podil:Q')
)
```

```python
mesta_z_wikipedie = [
    ['Pra[hz]','1 397 880'],
    ['Brn[oaeu]','402 739'],
    ['Ostrav','283 187'],
    ['Plz[en]','187 928'],
    ['Liber[ce]','108 090'],
    ['Olomou','103 063'],
    ['Buděj','97 231'],
    ['Hrad[ec]','	94 311'],
    ['Pardub','	92 319'],
    ['Ústí','90 866'],
    ['Zlín','74 684'],
    ['Klad[en]','69 664'],
    ['Havířo','68 674'],
    ['Opav','55 109'],
    ['Jihlav','54 624'],
    ['Frýd','	53 590'],
    ['Teplic','50 912'],
    ['Karvin','48 937'],
    ['Boleslav','47 346'],
    ['Chomutov','46 771'],
    ['Děčín','46 376'],
    ['Jablon','46 209'],
    ['Prostěj','43 408'],
    ['Přerov','40 906'],
    [r'Čes\w{1,5} Líp','36 815'],
    ['Třebíč','	34 530'],
    ['Znoj[em]','34 172'],
    ['Třin[ec]','33 852']
]
mesta = []

for m in mesta_z_wikipedie:
    mesta.append({'mesto':m[0],'obyvatel':int(m[1].replace(' ',''))})
```

```python
mesta
```

```python
soucasne_obdobi = df.filter(pl.col('obdobi') == 2021)
```

```python
statistiky_mest = []

for m in mesta:
    statistiky_mest.append(
        {'mesto':m['mesto'],'zminek':len(soucasne_obdobi.filter(pl.col('text').str.contains(m['mesto']))),'obyvatel':m['obyvatel']}
    )

pl.DataFrame(statistiky_mest).with_columns(
    (pl.col('obyvatel') / pl.col('zminek')).alias('pomer')
).sort(by='pomer')
```
