```python
import polars as pl
```

```python
from datetime import date
```

```python
pl.Config.set_tbl_rows(150)
pl.Config.set_fmt_str_lengths(150)
pl.Config.set_tbl_width_chars(200)
```

```python
df = pl.read_parquet("data/projevy.parquet").drop_nulls(subset='datum').sort(by="datum", descending=True)
```

```python
def naposledy(retezec, doslova=False):
    dfx = df.filter(pl.col('text').str.contains(retezec, literal=doslova)).head(1)
    if len(dfx) > 0:
        print(f"„{retezec}“ zmínil(a) naposledy {dfx.select(pl.col('mluvci')).item()} před {(date.today() - dfx.select(pl.col('datum')).item()).days} dny / {round((date.today() - dfx.select(pl.col('datum')).item()).days / 365.25, 2)} lety:")
        print("")
        print(f"{dfx.select(pl.col('text')).item()}")
    else:
        print('nenalezeno')
```
