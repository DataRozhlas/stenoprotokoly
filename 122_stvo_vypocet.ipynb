```python
import os
import string
import polars as pl
import altair as alt
```

```python
pl.Config.set_tbl_rows(300)
pl.Config.set_fmt_str_lengths(150)
pl.Config.set_tbl_width_chars(200)
```

```python
def bez_interpunkce(retezec):
    return retezec.translate(str.maketrans('', '', string.punctuation))
```

```python
bez_interpunkce("Jsem blbej? Ne, jen unavenej.")
```

```python
df = pl.read_parquet(
    os.path.join("data/snemovny_slova","snemovny_slova.parquet")
).with_columns(
    pl.col("text_merged").str.to_lowercase().map_elements(bez_interpunkce, return_dtype=pl.String)
)
```

```python
df.sample(3)
```

```python
stvo = df.with_columns(pl.col('text_merged').str.split(by=' ')).explode('text_merged').filter(pl.col('text_merged').str.ends_with("stvo"))
```
