```python
import os
import polars as pl
import altair as alt
```

```python
pl.Config.set_tbl_rows(150)
pl.Config.set_fmt_str_lengths(150)
pl.Config.set_tbl_width_chars(200)
```

```python
df = pl.read_parquet(os.path.join("data/snemovny_slova","snemovny_slova.parquet"))
```

```python
df
```

```python
def jak_casto(slovo):
    
    global df
    
    return df.with_columns(pl.col("text_merged").str.count_matches(slovo).alias("vyskytu")).with_columns((pl.col("vyskytu") / pl.col("slov")).alias("podil")).drop("text_merged")
```

```python
jak_casto("Brno")
```

```python
def graf(slovo):

    return alt.Chart(
        jak_casto(slovo)
    ).mark_bar(
    ).encode(
        alt.X("komora_pekna:N"),
        alt.Y("podil:Q")
    )
```

## Nejčastější nová slova

```python
slova_2021 = df.filter(pl.col("komora_pekna") == "2021_ps").select(pl.col("text_merged")).to_series().to_list()[0]
slova_2021 = ''.join([x.lower() for x in slova_2021 if x.isalnum() or x == ' ']).split(' ')
```

```python
from collections import defaultdict
```

```python
cetnost = defaultdict(int)
```

```python
for slovo in slova_2021:
    cetnost[slovo] += 1
```

```python
cetnost_zebricek = defaultdict(int, sorted(cetnost.items(), key=lambda x: x[1], reverse=True))
```

```python
cetnost_zebricek
```

```python
slova_starsi = df.filter(pl.col("komora_pekna") != "2021_ps").select(pl.col("text_merged")).to_series().to_list()
```

```python
from itertools import chain
```

```python
slova_starsi_flat1 = [x.split(' ') for x in slova_starsi]
```

```python
slova_starsi_flat2 = list(chain.from_iterable(slova_starsi_flat1))
```

```python
import re
```

```python
import itertools
```

```python
list(itertools.islice(slova_starsi_flat2, 10))
```

```python
slova_starsi_flat3 = [re.sub(r'[^a-zA-Z0-9ěščřžýáéíůúňďľťÉĚŠČŘŽÝÁÉÍÚŮŇĎĽŤóÓ]', '', word) for word in slova_starsi_flat2]
```

```python
list(itertools.islice(slova_starsi_flat3, 10))
```

```python
slova_starsi_flat = set([x.lower() for x in slova_starsi_flat3])
```

```python
len(slova_starsi_flat)
```

```python
for slovo, cetnost in cetnost_zebricek.items():
    if slovo not in slova_starsi_flat:
        print(f"{slovo} {cetnost}×")
```
