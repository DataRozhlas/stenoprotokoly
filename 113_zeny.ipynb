```python
import polars as pl
import altair as alt
```

```python
pl.Config.set_tbl_rows(150)
pl.Config.set_fmt_str_lengths(150)
pl.Config.set_tbl_width_chars(200)
```

```python
df = pl.read_parquet(
    "data/projevy.parquet"
).filter(
    pl.col('datum').dt.year() >= 1918
).drop_nulls(
    subset=['datum']
)
```

```python
df.filter(pl.col("text").str.contains("[Kk]am čert nemůže")).group_by_dynamic(index_column="datum",every="1y").agg(pl.col("text").unique().len())
```
